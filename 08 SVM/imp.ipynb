```python
import numpy as np 
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt 


cencer = load_breast_cancer()
X = cencer.data[:,:2]
y = cencer.target

plt.scatter(X[:, 0], X[:, 1], 
            c=y, 
            s=20, edgecolors="k")
plt.show()
plt.show()
```

```python
from sklearn.svm import SVC

model = SVC(kernel='poly', degree=5)
model.fit(X, y)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

DecisionBoundaryDisplay.from_estimator(
        model,
        X,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=cencer.feature_names[0],
        ylabel=cencer.feature_names[1],
    )

# Scatter plot
plt.scatter(X[:, 0], X[:, 1], 
            c=y, 
            s=20, edgecolors="k")
plt.show()
```
